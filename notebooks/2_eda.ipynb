```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/Users/unnatir/Desktop/ML/Food_Wastage_Prediction/data/raw/food_wastage_data.csv")
df.head()
```

```python
df=df.rename(columns={
    'Type of Food':'food_type',
    'Number of Guests':'num_guests',
    'Quantity of Food':'quantity_prepared',
    'Wastage Food Amount':'quantity_wasted',
    'Event Type':'event_type'
})
```

```python
df.describe()
```

```python
plt.Figure(figsize=(6,4))
sns.barplot(x='food_type',y='quantity_wasted',data=df,estimator=sum)
plt.title('Food_Type VS Quantity Wasted')
plt.show()
```

```python
plt.Figure(figsize=(6,4))
sns.scatterplot(x='num_guests',y='quantity_wasted',hue='event_type',data=df)
plt.title('Number of Guests VS Quantity Wasted')
plt.show()
```

```python
plt.Figure(figsize=(6,4))
sns.scatterplot(x='quantity_prepared',y='quantity_wasted',data=df)
sns.regplot(x='quantity_prepared',y='quantity_wasted',data=df,scatter=False,color='red')
plt.title('Quantity Prepared VS Quantity Wasted')
plt.show()
```

```python
plt.Figure(figsize=(6,4))
sns.boxplot(x='event_type',y='quantity_wasted',data=df)
plt.title('Event Type VS Qunatity Wasted')
plt.show()
```
